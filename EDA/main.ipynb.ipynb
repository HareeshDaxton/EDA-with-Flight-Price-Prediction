```python
import pandas as pd    
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn                                                      
```

```python
df = pd.read_excel('Data_Train.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype('int64')
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype('int64')
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype('int64')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0].astype('int64')
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1].astype('int64')
```

```python
df.head()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0].astype('int64')
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype('int64')
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['Duration'] = df['Duration'].fillna('0h 0m')
df['Duration_hour'] = df['Duration'].apply(lambda x: int(x.split()[0].replace('h', '')) if 'h' in x.split()[0] else 0).astype('int64')
df['Duration_min'] = df['Duration'].apply(lambda x: int(x.split()[-1].replace('m', '')) if 'm' in x.split()[-1] else 0).astype('int64')
```

```python
df.drop('Duration', axis=1, inplace=True)   
df.head()
```
